```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(18,18))

sns.heatmap(
    df.corr(numeric_only=True),
    ax=ax,
    annot=True, #per vedere i numeri oltre che il mero grafico
    fmt='.2f', #decimali considerati
    cmap=sns.color_palette('magma'), #palette preferita
    linewidths=1 #linea separazione
    )

plt.show()
```

```python
#vediamo infatti che così facendo prendiamo in esame soltanto questa serie del dataframe
print(type(df.corr(numeric_only=True)['popularity']))
df.corr(numeric_only=True)['popularity']
```

```python
#prendiamo i valori assoluti della serie e ordiniamoli
serie = np.abs(df.corr(numeric_only=True)['popularity']).sort_values(ascending=False)
serie
```

```python
# disegno un histogramma PER OGNI feature numerica in modo da
# rendermi conto della distribuzione dei dati

features_numeriche = df.select_dtypes(include=np.number)

for col in features_numeriche.columns:
    sns.histplot(features_numeriche[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
artisti_piu_popolari = df.groupby('artists')['popularity'].sum().sort_values(ascending=False)

top_20_artisti = artisti_piu_popolari.head(20)

sns.barplot(
    x=top_20_artisti.values,
    y=top_20_artisti.index,
    hue=top_20_artisti.index,
    palette="Spectral"
)

plt.title('Top 20 Artisti più Popolari')
plt.ylabel(None) #inutile lasciarla di default
plt.show()
```

```python
# prendo solo la top 10 di artisti con più di 100 canzoni
df['popular_artist'] = df['artists'].map(df['artists'].value_counts() > 100) 
artisti_big = df.loc[df['popular_artist'] == True].copy()

artisti_big = artisti_big.groupby('artists')['popularity'].mean().sort_values(ascending=False)

print('Ecco i nomi',artisti_big.head(10).index)
print('Ecco le loro popolarità',artisti_big.head(10).values)
```

```python
sns.barplot(
    x=artisti_big.head(10).values,
    y=artisti_big.head(10).index,
    hue=artisti_big.head(10).index,
    palette="Spectral"
)

plt.title('Top 10 Artisti BIG più Popolari')
plt.ylabel(None) #inutile lasciarla di default
plt.show()
```
